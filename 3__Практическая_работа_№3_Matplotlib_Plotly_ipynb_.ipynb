## Комплексное задание №1: Моделирование и визуализация распространения инфекционных болезней в популяции с использованием ООП подхода

- Разработать и реализовать модель SIR для анализа динамики распространения инфекционной болезни в закрытой популяции. Применить принципы объектно-ориентированного программирования (ООП) для структурирования кода, использовать библиотеку NumPy для математических вычислений и библиотеку Plotly Express для создания интерактивных визуализаций результатов моделирования.



#### Описание модели SIR:

Модель SIR (Susceptible, Infected, Recovered) представляет собой классическую эпидемиологическую модель, которая разделяет популяцию на три основные группы:
- **S (Susceptible)**: Часть популяции, восприимчивая к инфекции, но еще не зараженная.
- **I (Infected)**: Часть популяции, зараженная инфекцией и способная передавать ее другим.
- **R (Recovered)**: Часть популяции, выздоровевшая после инфекции и получившая иммунитет.

#### Математический аппарат модели SIR:

1. **Дифференциальные уравнения модели SIR:**

   - $\frac{dS}{dt} = -\beta \frac{SI}{N}$ — уменьшение числа восприимчивых из-за заражения.
   - $\frac{dI}{dt} = \beta \frac{SI}{N} - \gamma I$ — изменение числа инфицированных, учитывая новые случаи заражения и выздоровление.
   - $\frac{dR}{dt} = \gamma I$ — увеличение числа выздоровевших благодаря выздоровлению инфицированных.

   Где:
   - $N$ — общая численность популяции (постоянная величина).
   - $\beta$ — коэффициент заразности, определяющий вероятность передачи инфекции от инфицированных к восприимчивым.
   - $\gamma$ — коэффициент выздоровления, обратный среднему периоду инфекционности.

2. **Задание:**
   - Создайте класс `SIRModel`, инкапсулирующий параметры и методы модели. Класс должен включать методы для инициализации параметров, вычисления изменений во времени и сохранения результатов моделирования.
   - Примените библиотеку NumPy для выполнения вычислений и хранения результатов в виде массивов.
   - Разработайте метод для визуализации динамики групп популяций (S, I, R) с течением времени, используя интерактивные возможности Plotly Express  для наглядного представления результатов.

3. **Структура класса:**

  Класс `SIRModel` предназначен для моделирования распространения инфекционных заболеваний в популяции и включает в себя следующие компоненты:

  **Свойства:**
  - `N`: Общая численность популяции, представляет собой постоянное значение для всего периода моделирования.
  - `beta`: Коэффициент передачи инфекции, определяет скорость, с которой болезнь передается от инфицированных к восприимчивым.
  - `gamma`: Скорость выздоровления, указывает на частоту, с которой инфицированные особи переходят в состояние выздоровевших.
  - `S`: Количество восприимчивых к болезни в начале моделирования.
  - `I`: Количество инфицированных в начале моделирования.
  - `R`: Количество выздоровевших в начале моделирования.
  - `t_max`: Максимальное время моделирования, определяет продолжительность наблюдения за динамикой распространения болезни.
  - `dt`: Шаг по времени, используется для дискретизации времени при моделировании.
  - `t`: Текущее время моделирования, начинается с нуля и увеличивается на `dt` с каждым шагом.
  - `S_list`, `I_list`, `R_list`, `t_list`: Списки, хранящие исторические данные о количестве восприимчивых, инфицированных, выздоровевших и времени соответственно.

  **Методы:**
  - `__init__`: Конструктор класса, инициализирует все параметры модели и задает начальные условия.
  - `run`: Основной метод, выполняющий моделирование. Использует дифференциальные уравнения для расчета изменений в количестве восприимчивых, инфицированных и выздоровевших на каждом шаге времени.
  - `plot`: Метод для визуализации результатов моделирования. Создает интерактивный график, отображающий динамику изменения популяций во времени.

  **Характеристика методов:**
  - **Метод `run`** использует численные методы для решения системы дифференциальных уравнений, что позволяет проследить за изменениями в количестве восприимчивых, инфицированных и выздоровевших в зависимости от времени.

    В методе `run` класса `SIRModel`, используются следующие формулы для вычисления изменений количества восприимчивых (S), инфицированных (I) и выздоровевших (R) в популяции:

  - Изменение восприимчивых:
    $$\Delta S = -\beta \frac{SI}{N} \Delta t$$

  - Изменение инфицированных:
    $$\Delta I = \left(\beta \frac{SI}{N} - \gamma I\right) \Delta t$$

  - Изменение выздоровевших:
    $$\Delta R = \gamma I \Delta t$$

    Где:
    - $ \beta $ — коэффициент передачи инфекции, который определяет вероятность заражения.
    - $ \gamma $ — скорость выздоровления, которая указывает на частоту перехода инфицированных в состояние выздоровевших.
    - $ N $ — общая численность популяции.
    - $ \Delta t $ — шаг по времени.

    Эти формулы применяются на каждом шаге времени в цикле моделирования для обновления значений $ S $, $ I $, и $ R $.

  - **Метод `plot`** применяет библиотеку Plotly Express для создания графика, что обеспечивает наглядность и интерактивность представления данных. Пользователь может взаимодействовать с графиком, например, изменяя масштаб или просматривая точные значения популяций в различные моменты времени.

```python
class SIRModel:
    def __init__(self, N, beta, gamma, S0, I0, R0, t_max, dt):
        self.N = N
        self.beta = beta
        self.gamma = gamma
        self.S = np.array([S0])
        self.I = np.array([I0])
        self.R = np.array([R0])
        self.t = np.array([0])
        self.t_max = t_max
        self.dt = dt

    def run(self):
        while self.t[-1] < self.t_max:
            dS = -self.beta * self.S[-1] * self.I[-1] / self.N * self.dt
            dI = (self.beta * self.S[-1] * self.I[-1] / self.N - self.gamma * self.I[-1]) * self.dt
            dR = self.gamma * self.I[-1] * self.dt

            self.S = np.append(self.S, self.S[-1] + dS)
            self.I = np.append(self.I, self.I[-1] + dI)
            self.R = np.append(self.R, self.R[-1] + dR)
            self.t = np.append(self.t, self.t[-1] + self.dt)

    def plot(self):
        fig = px.line(x = self.t, y = [self.S, self.I, self.R])
        fig.update_layout(
            title='Динамика распространения заболенвания',
            xaxis_title='Время, дни',
            yaxis_title='Количество людей',
            legend_title='Группа популяции',
            legend=dict(
              bgcolor="#B0C4DE",
              bordercolor="Black",
              borderwidth=2,
              traceorder='normal'
              ),
            legend_traceorder='normal'
            )

        for i in range(len(fig.data)):
          fig.data[i].name = ['Восприимчивые (S)', 'Зараженные (I)', 'Выздоровевшие (R)'][i]
          fig.data[i].line.width = 2

        fig.show()
```

```python
# Создаем объект модели SIR с заданными параметрами
model = SIRModel(N=1000, beta=0.2, gamma=0.1, S0=999, I0=1, R0=0, t_max=200, dt=0.1)
# Запускаем моделирование
model.run()
# Визуализируем результаты
model.plot()
```

4. **Особенности математического аппарата**

- **Класс `Prey` (Жертвы):**
  1. **Рост популяции жертв:** Прирост популяции жертв без учета хищников описывается уравнением $$ dx = \alpha x $$ где $ \alpha $ - коэффициент роста, $ x $ - текущее количество жертв.
  2. **Убыль популяции жертв:** Убыль популяции жертв из-за хищников описывается уравнением $$ dx = -\beta x y $$ где $ \beta $ - коэффициент смертности жертв от хищников, $ y $ - количество хищников.

- **Класс `Predator` (Хищники):**
  1. **Рост популяции хищников:** Прирост популяции хищников за счет питания жертвами описывается уравнением $$ dy = \delta x y $$ где $ \delta $ - коэффициент эффективности превращения пищи в новых хищников.
  2. **Убыль популяции хищников:** Убыль популяции хищников из-за естественной смертности описывается уравнением $$ dy = -\gamma y $$ где $ \gamma $ - коэффициент естественной смертности хищников.

- **Класс `EcoSystem` (Экосистема):**
  - В этом классе происходит моделирование взаимодействия между хищниками и жертвами. Используя метод `run`, модель выполняет итерации, на каждом шаге рассчитывая изменения в популяциях с помощью уравнений, описанных выше, и обновляя их значения.


Ваша реализация:

```python
# Импортируем необходимые библиотеки
import numpy as np
import plotly.express as px

# Создаем класс для моделирования популяции жертв
class Prey:
    def __init__(self, x0, alpha):
        # Инициализируем начальное количество жертв и коэффициент роста
        self.x = x0
        self.alpha = alpha

    def grow(self):
        # Рассчитываем прирост популяции жертв без учета хищников
        return self.alpha * self.x

    def die(self, y, beta):
        # Рассчитываем убыль популяции жертв из-за хищников
        return beta * self.x * y

    def update(self, dx):
        # Обновляем количество жертв с учетом прироста и убыли
        self.x += dx

# Создаем класс для моделирования популяции хищников
class Predator:
    def __init__(self, y0, gamma):
        # Инициализируем начальное количество хищников и коэффициент смертности
        self.y = y0
        self.gamma = gamma

    def grow(self, x, delta):
        # Рассчитываем прирост популяции хищников за счет питания жертвами
        return delta * x * self.y

    def die(self):
        # Рассчитываем убыль популяции хищников из-за естественной смертности
        return self.gamma * self.y

    def update(self, dy):
        # Обновляем количество хищников с учетом прироста и убыли
        self.y += dy

# Создаем класс для моделирования динамики популяции хищник-жертва
class EcoSystem:
    def __init__(self, x0, y0, alpha, beta, gamma, delta, t_max, dt):
        # Инициализируем параметры модели
        self.prey = Prey(x0, alpha)
        self.predator = Predator(y0, gamma)
        self.beta = beta
        self.delta = delta
        self.t_max = t_max
        self.dt = dt
        self.t = 0
        self.x_list = [self.prey.x]
        self.y_list = [self.predator.y]
        self.t_list = [self.t]

    def run(self):
        # Запускаем моделирование
        while self.t < self.t_max:
            dx = self.prey.grow() - self.prey.die(self.predator.y, self.beta)
            dy = self.predator.grow(self.prey.x, self.delta) - self.predator.die()
            self.prey.update(dx * self.dt)
            self.predator.update(dy * self.dt)
            self.t += self.dt
            self.x_list.append(self.prey.x)
            self.y_list.append(self.predator.y)
            self.t_list.append(self.t)

    def plot(self):
        x_arr = np.array(self.x_list)
        y_arr = np.array(self.y_list)
        t_arr = np.array(self.t_list)

        df = {
            'Время':np.array(self.t_list),
            'Жертвы': np.array(self.x_list),
            'Хищники': np.array(self.y_list)
        }

        fig = px.line(df, x='Время', y=['Жертвы', 'Хищники'], title='Динамика популяций хищник-жертва во времени')
        fig.update_layout(
            legend_title_text='Виды популяций',
            xaxis_title='Время, единицы',
            yaxis_title='Количество особей',
            )

        fig_phase = px.scatter(x=x_arr, y=y_arr, title='Фазовый портрет системы хищник-жертва',
                               labels={'x': 'Популяция жертв', 'y': 'Популяция хищников'})
        fig.show()
        fig_phase.show()
```

```python
# Создаем объект экосистемы с заданными параметрами
ecosystem = EcoSystem(x0=10, y0=5, alpha=0.1, beta=0.01, gamma=0.1, delta=0.01, t_max=1000, dt=0.1)
# Запускаем моделирование
ecosystem.run()
# Визуализируем результаты
ecosystem.plot()
```
